## 1. Setup

Import the pipeline and create a configuration.

```python
from pyTransitPhotometry import TransitPipeline, PipelineConfig
from pyTransitPhotometry.config import create_example_config
import matplotlib.pyplot as plt
import numpy as np

# Set matplotlib style
plt.style.use('seaborn-v0_8-darkgrid')
```

## 2. Create Configuration

First time? Create an example config file:

```python
# Create example configuration (only needed once)
create_example_config('my_config.yaml')

# Edit my_config.yaml to match your data!
# Then load it:
config = PipelineConfig.from_yaml('config_example.yaml')

# View configuration summary
config.summary()
```

## 3. Run Complete Pipeline

### Option A: Run everything at once

```python
# Create pipeline
pipeline = TransitPipeline(config)

# Run all stages
results = pipeline.run()

# Access results
print(f"\nTransit depth: {results['fit_result']['derived_params']['transit_depth_pct'][0]:.2f}%")
print(f"Impact parameter: {results['fit_result']['derived_params']['impact_parameter'][0]:.3f}")
```

### Option B: Run stage-by-stage (for more control)

```python
# Create pipeline
pipeline = TransitPipeline(config)

# Stage 1: Calibration
pipeline.run_calibration()
print(f"Calibrated {len(pipeline.calibrated_images)} frames")
```

```python
# Stage 2: Detection
pipeline.run_detection()
print(f"Detected {len(pipeline.sources_list[0])} stars in first frame")
```

```python
# Stage 3: Photometry
pipeline.run_photometry()
print(f"Extracted light curve: {len(pipeline.lightcurve['times'])} points")
```

```python
# Stage 4: Detrending
pipeline.run_detrending()
print(f"After detrending: {len(pipeline.detrended_lc['times'])} points")
```

```python
# Stage 5: Transit Fitting
pipeline.run_transit_fit()

# Print results
fit = pipeline.fit_result['fitted_params']
print(f"\nRp/Rs = {fit['rp'][0]:.4f} ± {fit['rp'][1]:.4f}")
print(f"a/Rs = {fit['a'][0]:.2f} ± {fit['a'][1]:.2f}")
print(f"inc = {fit['inc'][0]:.2f} ± {fit['inc'][1]:.2f} degrees")
```

```python
# Stage 6: Export
pipeline.export_results()
print("Results saved to", config.paths.output_dir)
```

## 4. Visualizations

Use the visualization module for publication-quality plots.

```python
from transitphotometry.visualization import (
    plot_calibration_comparison,
    plot_detected_sources,
    plot_lightcurve,
    plot_transit_fit
)
```

### Calibration Before/After

```python
plot_calibration_comparison(
    pipeline.science_data[0],
    pipeline.calibrated_images[0],
    save_path='outputs/calibration.png'
)
```

### Detected Stars

```python
plot_detected_sources(
    pipeline.calibrated_images[0],
    pipeline.sources_list[0],
    target_index=config.photometry.target_star_index,
    reference_indices=config.photometry.reference_star_indices,
    save_path='outputs/sources.png'
)
```

### Raw Light Curve

```python
plot_lightcurve(
    pipeline.lightcurve['times'],
    pipeline.lightcurve['fluxes'],
    pipeline.lightcurve['errors'],
    title='Raw Differential Photometry',
    save_path='outputs/lightcurve_raw.png'
)
```

### Transit Fit

```python
# Generate model light curve for plotting
model_times = np.linspace(
    pipeline.detrended_lc['times'].min(),
    pipeline.detrended_lc['times'].max(),
    500
)

plot_transit_fit(
    pipeline.detrended_lc['times'],
    pipeline.detrended_lc['fluxes'],
    pipeline.detrended_lc['errors'],
    model_times,
    pipeline.fit_result['model_flux'],
    pipeline.fit_result['residuals'],
    pipeline.fit_result['fitted_params'],
    save_path='outputs/transit_fit.png'
)
```

## 5. Access Individual Components

Use individual modules for custom workflows.

```python
# Example: Custom photometry on a single image
from pyTransitPhotometry.photometry import measure_flux
from pyTransitPhotometry.io import get_ccd_gain

# Get CCD gain
ccd_gain = get_ccd_gain(pipeline.headers[0])

# Measure flux at a position
result = measure_flux(
    pipeline.calibrated_images[0],
    position=(512, 512),  # x, y in pixels
    aperture_radius=8.0,
    annulus_inner=40.0,
    annulus_outer=60.0,
    ccd_gain=ccd_gain
)

print(f"Flux: {result['flux']:.1f} ± {result['flux_err']:.1f}")
print(f"SNR: {result['snr']:.1f}")
print(f"Background: {result['background_mean']:.2f} counts/pixel")
```

## 6. Batch Processing Multiple Targets

Process multiple targets or nights easily.

```python
targets = [
    {'config': 'wasp75b_night1.yaml', 'name': 'WASP-75b Night 1'},
    {'config': 'wasp75b_night2.yaml', 'name': 'WASP-75b Night 2'},
]

all_results = []

for target in targets:
    print(f"\nProcessing {target['name']}...")
    
    config = PipelineConfig.from_yaml(target['config'])
    pipeline = TransitPipeline(config)
    results = pipeline.run()
    
    all_results.append({
        'name': target['name'],
        'rp_rs': results['fit_result']['fitted_params']['rp'],
        'depth': results['fit_result']['derived_params']['transit_depth_pct'],
        'chi2': results['fit_result']['reduced_chi_squared']
    })

# Compare results
import pandas as pd
df = pd.DataFrame(all_results)
print("\nSummary:")
print(df)
```

## 7. Export Results

Results are automatically saved to the output directory:

- `lightcurve_raw.csv`: Raw differential photometry
- `lightcurve_detrended.csv`: After sigma clipping
- `fit_results.json`: Transit parameters with uncertainties
- `config_used.yaml`: Configuration for reproducibility

Load saved results:

```python
import pandas as pd
import json

# Load light curve
lc = pd.read_csv('outputs/lightcurve_detrended.csv')
print(f"Light curve: {len(lc)} points")

# Load fit results
with open('outputs/fit_results.json') as f:
    fit_data = json.load(f)

print(f"\nRp/Rs = {fit_data['parameters']['rp']['value']:.4f} ± {fit_data['parameters']['rp']['uncertainty']:.4f}")
```
